<h4>c. For the ward from part b, plot the trends between the number of crimes occurring
in this ward along with the unemployment rate of this ward based on the years
chronologically present in the datasets. Note that you may have to clean the data
to exclude some years which may not have data for the entire year, as this will
skew your analysis. Explain your procedure and what you observed. ( 5 points )</h4?

```python
index_max_warden_unemployment = df2_unemployment_avg_sum['Unemployment Rate'].idxmax()
df_filtered_max_unemp_avg = df[df['WARD'] == index_max_warden_unemployment]

years_indices = df_filtered_max_unemp_avg.groupby('year').count().index
valid_2c_years = []

for i in years_indices:
    if(df_filtered_max_unemp_avg[df_filtered_max_unemp_avg['year'] == i].groupby('month').count().index.size == 12):
        valid_2c_years.append(i)

df2_filtered_max_unemp_avg = df_filtered_max_unemp_avg[df_filtered_max_unemp_avg['year'].isin(valid_2c_years)]

part2c_numreport = df2_filtered_max_unemp_avg.groupby('year').size()
part2c_unemploy = df2[(df2['Ward'] == index_max_warden_unemployment) & (df2['Year'].isin(valid_2c_years))]

part2c_combined = pd.merge(part2c_numreport.to_frame(), part2c_unemploy[['Year', 'Unemployment Rate']], left_on='year', right_on='Year')
part2c_combined.columns = ['Reported Cases', 'Year', 'Unemployment Rate']
```

```python
plt.plot(part2c_combined['Year'], part2c_combined['Reported Cases'])
plt.xlabel('Year')
plt.ylabel('Number of Reported Cases per Year')
plt.title('Number of Reported Cases per Year for the Ward the highest Average Unemployment Rate')
plt.show()
```

```python
plt.plot(part2c_combined['Year'], part2c_combined['Unemployment Rate'])
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate per Year for the Ward the highest Average Unemployment Rate')
plt.show()
```

<h5>To remove the years that did not contain all of the data involving criminal reports, I first narrowed my data down so that it would only contain the ward that reported the highest average unemployment rates. I created one list containing all of the years present in the dataset and then grouped the months for each year together. If the number of months was not equal to 12 (the total number of months in a year), I did not add it to a list that was originally empty. I narrowed down my data to only contain the years in this list (2017 was excluded because it only contained 11 months in the dataset). I grouped the data by year once again and stored the number of rows (reported cases) into a series. I removed any years (2017) that did not contain sufficient data from the other dataset as well. I merged both datasets but used the options "left_on" and "right_on" because of case sensitivity. I then created the two plots above based off of that data frame.</h5>

<h5>I observed that the unemployment rate and the number of reported cases had an overall downward trend from 2008 to 2016.</h5>

<h5>In the plot above, I wanted to see the trend of crimes involving guns in the area. I filtered out the year 2017 once again as it would skew the plot. I wanted to see if President Barack Obama's implementation of gun control laws would decrease the number of total crimes involving guns. It was interesting to see that there was an overall downward trend on gun violence in the duration of his presidency.
```python
plt.figure(figsize=(10, 8))
plt.rcParams.update({'font.size': 9})

df_method_guns = df[df['METHOD'] == 'GUN']
df_method_guns_a = df_method_guns.groupby('OFFENSE').size()

df_method_knives = df[df['METHOD'] == 'KNIFE']
df_method_knives_a = df_method_knives.groupby('OFFENSE').size()

plt.bar(df_method_guns_a.index, df_method_guns_a.values, color='r', label='Guns')
plt.bar(df_method_knives_a.index, df_method_knives_a.values, bottom=df_method_guns_a.values, color='g', label='Knives')

plt.xticks(rotation=45, ha='right')
plt.legend(title="Method")
plt.xlabel('Offense Type')
plt.ylabel('Frequency')
plt.title('Number of Reported Crimes per Offense Type Catergorized by Method')
plt.show()
```

<h5>In the bar graph above, I show the relationship between the offense type and the method used for the offense.
